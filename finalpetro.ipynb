```python
list_of_times = []
for i in range(len(df_old)):
# for i in range(2):
    missed_timestamps = 0
    wrong_time = 0
    list_of_times = list_of_times + str(df_old['time'][i]).split(', ')
#     print("list_of_times :",list_of_times)
#     list_of_times[0] = list_of_times[0][1:]
#     print("list_of_times[-1] :",list_of_times[-1])    
#     list_of_times[-1] = list_of_times[-1][:-1]
#     print("list_of_times[-1] :",list_of_times[-1])
    
    for j in range(len(list_of_times)):
#         print(len(list_of_times))
        if (j+1 != len(list_of_times)):
            list_of_times[j] = list_of_times[j].strip("'")
            list_of_times[j+1] = list_of_times[j+1].strip("'")
            if list_of_times[j][0].isdigit() and list_of_times[j+1][0].isdigit():
                t_now = datetime.strptime(list_of_times[j], "%H:%M")
                t_next = datetime.strptime(list_of_times[j+1], "%H:%M")
                

                if t_next - t_now == timedelta(hours=0, minutes=5):
                    pass
                else:
                    wrong_time = wrong_time + 1
            else:
                missed_timestamps = missed_timestamps + 1
                
# #     print(wrong_time)
#     print(missed_timestamps)

    if (wrong_time > 0) or (missed_timestamps > 0):
        print("time check failed. There were " + str(wrong_time) + " instances of a timestamp at the wrong time and " + str(missed_timestamps) + " instances of a timestamp that was missed")

```
